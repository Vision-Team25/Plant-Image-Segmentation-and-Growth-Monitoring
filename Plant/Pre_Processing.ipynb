```python
# OpenCV library for image processing
import cv2  
# NumPy for numerical operations
import numpy as np 
# Matplotlib for displaying images
import matplotlib.pyplot as plt  
```

```python
# Load the plant image
# Replace "plant.jpg" with the path to your image file
image = cv2.imread("plant.jpg")

# Convert the image from BGR format to RGB format
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image using Matplotlib
 # Set the figure size to 6x6 inches
plt.figure(figsize=(6, 6)) 
# Display the image
plt.imshow(image)  
# Set the title of the image
plt.title("Original Image")  
# Hide the axis
plt.axis("off") 
# Show the image
plt.show()
```

```python
# Convert the RGB image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

# Display the grayscale image using Matplotlib
# Set the figure size to 6x6 inches
plt.figure(figsize=(6, 6))
# Display the image in grayscale with the "gray" colormap
plt.imshow(gray, cmap="gray")
# Set the title of the image
plt.title("Grayscale Image")
# Hide the axis
plt.axis("off")
# Show the image
plt.show()
```

```python
# Apply Gaussian Blur to the grayscale image to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Display blurred image using Matplotlib
 # Set the figure size to 6x6 inches
plt.figure(figsize=(6, 6))
# Display the image in grayscale with the "gray" colormap
plt.imshow(blurred, cmap="gray")
# Set the title of the image
plt.title("Noise Reduction - Gaussian Blur")
# Hide the axis
plt.axis("off")
# Show the image
plt.show()
```

```python
# Apply CLAHE (Contrast Limited Adaptive Histogram Equalization) to enhance contrast

# CLAHE is useful for improving the visibility of image details in regions with varying illumination
clahe = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8, 8))
# Apply the CLAHE algorithm to the blurred grayscale image
equalized = clahe.apply(blurred)

# Display the contrast-enhanced image using Matplotlib
# Set the figure size to 6x6 inches
plt.figure(figsize=(6, 6))
# Display the image in grayscale with the "gray" colormap
plt.imshow(equalized, cmap="gray")
# Set the title of the image
plt.title("Contrast Enhancement - CLAHE")
# Hide the axis
plt.axis("off")
# Show the image
plt.show()
```

```python
# Apply Otsu's thresholding for background subtraction

# Otsu's method automatically determines the optimal threshold value to convert the grayscale image to a binary image
# In this case, equalized is the input image, and binary_mask is the output binary image
# cv2.THRESH_BINARY + cv2.THRESH_OTSU specifies that binary thresholding and Otsu's method are used together
_, binary_mask = cv2.threshold(equalized, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the binary mask (thresholded image) using Matplotlib
# Set the figure size to 6x6 inches
plt.figure(figsize=(6, 6))
# Display the image in grayscale with the "gray" colormap
plt.imshow(binary_mask, cmap="gray")
# Set the title of the image
plt.title("Background Subtraction - Otsu's Thresholding")
# Hide the axis
plt.axis("off")
# Show the image
plt.show()
```

```python
# Apply all steps together: Noise Reduction → Contrast Enhancement → Otsu’s Thresholding

# Apply Gaussian Blur to the grayscale image to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Apply CLAHE (Contrast Limited Adaptive Histogram Equalization) to enhance contrast
equalized = clahe.apply(blurred)

# Apply Otsu's thresholding for background subtraction
_, final_mask = cv2.threshold(equalized, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Display the final pre-processed image using Matplotlib
# Set the figure size to 6x6 inches
plt.figure(figsize=(6, 6))
# Display the image in grayscale with the "gray" colormap
plt.imshow(final_mask, cmap="gray")
# Set the title of the image
plt.title("Final Pre-Processed Image with Otsu’s Thresholding")
# Hide the axis
plt.axis("off")
# Show the image
plt.show()
```
